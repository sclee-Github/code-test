```python
# Ex. 4-1 (상하좌우)
n = int(input())
direction = input().split()

x, y = 1, 1

dx = [0, 0, 1, -1]
dy = [1, -1, 0, 0]

move_types = ['R', 'L', 'D', 'U']

for s in direction:
    if s == 'R':
        nx = x + dx[0]
        ny = y + dy[0]

    elif s == 'L':
        nx = x + dx[1]
        ny = y + dy[1]
    
    elif s == 'D':
        nx = x + dx[2]
        ny = y + dy[2]

    elif s == 'U':
        nx = x + dx[3]
        ny = y + dy[3]
    
    # for type in range(len(move_types)):
    #     if s == move_types[type]:
    #         nx = x + dx[type]
    #         ny = y + dy[type]

    if nx < 1 or ny < 1 or nx > n or ny > n:
        continue

    x, y = nx, ny

print(x, y)
```

```python
# Ex. 4-2 (시각)
n = int(input())

count = 0

for hour in range(0, n+1):
    for minute in range(0, 60):
        for second in range(0, 60):
            # if '3' in str(hour) or '3' in str(minute) or '3' in str(second):            
            if '3' in str(hour) + str(minute) + str(second):                            
                count += 1

print(count)
                
```

```python
# Ex. 4-3 (왕실의 나이트)

coordinate = input()

row = int(coordinate[1])
column = int(ord(coordinate[0])) - int(ord('a')) + 1

move_types = [(2, 1), (2, -1), (-2, -1), (-2, 1), (-1, 2), (1, 2), (-1, -2), (1, -2)]
count = 0

for type in move_types:
    newt_row = row + type[0]
    newt_column = column + type[1]

    if newt_row >= 1 and newt_column >= 1 and newt_row <= 8 and newt_column <= 8:
        count += 1

print(count)
```
