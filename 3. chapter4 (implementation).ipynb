```python
# Ex. 4-3 (왕실의 나이트)

coordinate = input()

row = int(coordinate[1])
column = int(ord(coordinate[0])) - int(ord('a')) + 1

move_types = [(2, 1), (2, -1), (-2, -1), (-2, 1), (-1, 2), (1, 2), (-1, -2), (1, -2)]
count = 0

for type in move_types:
    newt_row = row + type[0]
    newt_column = column + type[1]

    if newt_row >= 1 and newt_column >= 1 and newt_row <= 8 and newt_column <= 8:
        count += 1

print(count)
```

```python
# Ex. 4-4 (게임 개발)
n, m = map(int, input().split())

# 방문한 곳 count를 위한 지도
arr_count = [[0] * m for _ in range(n)]

# 좌표 및 방향 입력 받기 (0: 북쪽, 1: 동쪽, 2: 남쪽, 3: 서쪽)
x, y, direction = map(int, input().split())
arr_count[x][y] = 1
count = 1
turn_time = 0

# 바다, 육지 지도 입력
arr_game = []
for i in range(n):
    arr_game.append(list(map(int, input().split())))

dx = [-1, 0, 1, 0] # 북, 동, 남, 서
dy = [0, 1, 0, -1]

def turn_left():
    global direction
    direction -= 1
    if direction == -1:
        direction = 3
    
while True:
    turn_left()

    nx = x + dx[direction]
    ny = y + dy[direction]

    if nx < 0 or ny < 0 or nx >= n or ny >= m:
        turn_time += 1    
    elif arr_game[nx][ny] == 0 and arr_count[nx][ny] == 0:
        x = nx
        y = ny
        arr_count[nx][ny] = 1
        count += 1
        turn_time = 0        
        continue
    else:
        turn_time += 1        

    if turn_time == 4:
        nx = x - dx[direction]
        ny = y - dy[direction]

        if nx >= 0 and ny >= 0 and nx < n and ny < m:
            if arr_game[nx][ny] == 1:
                break
            else:
                x = nx
                y = ny                                
        else:
            break

        turn_time = 0        

print(count)
```

```python
# Baekjoon (#1316 그룹 단어 체커)
n = int(input())

list_word = []

count = 0

for _ in range(n):
    list_word.append(input())

for word in list_word:
    b_group = True

    for i in range(len(word)-1):         
        if word[i] != word[i+1]:
            new_word = word[i+1:]            

            if new_word.count(word[i]) >= 1:
                b_group = False
    
    if b_group == True:        
        count += 1

print(count)
```
