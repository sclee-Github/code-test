```python
# Ex. 3-1 (거스름돈)
n = 1260
count = 0

list_change = [500, 100, 50, 10]

for coin in list_change:
    count += n // coin
    n %= coin

print(count)
```

```python
# Ex. 3-3 (숫자 카드 게임)
n, m = map(int, input().split())

list_min = []

for i in range(n):
    data = list(map(int, input().split()))
    list_min.append(min(data))

res = max(list_min)

print(res)
```

```python
# Ex. 3-6 (1이 될 때까지)

n, k = map(int, input().split())

# region answer code
result = 0

while True:
    target = (n // k) * k # n과 가장 가까운 k 배수 찾기
    result += (n - target) # -1 연산 횟수 추가
    n = target 

    if (n < k):
        break

    n //= k
    result += 1

result += (n-1)
print(result)
# endregion

# region my code
# result = 0

# while(n != 1):
#     if(n % k == 0):        
#         n = n // k
#         result += 1
#     else:
#         n -= 1
#         result += 1        

# print(result)
# endregion
```

```python
# 유튜브 강의 Ex (곱하기 혹은 더하기)
S = input()

result = int(S[0])

for i in range(1, len(S)):
    num = int(S[i])

    if num <= 1 or result <= 1:
        result += num
    else:
        result *= num
    

print(result)
```

```python
# 유튜브 강의 Ex (모험가 길드)
n = int(input())
list_nums = list(map(int, input().split()))
list_nums.sort()

result = 0 # 총 그룹의 수
count = 0  # 현재 그룹의 모험가의 수

for i in range(len(list_nums)):
    count += 1

    if count >= list_nums[i]:
        result += 1
        count = 0

print(result)
```

```python
# Baekjoon (#1789 수들의 합)

n = int(input())

sum = 0
res = 0

for i in range(1, n+1):
    sum += i
    res += 1

    if sum > n:
        res -= 1
        break

print(res)
```

```python
# Baekjoon (#2839 설탕 배달)
num = int(input())
count = 0

while num >= 0:
    if num % 5 == 0:
        count += int(num // 5)
        print(count)
        break

    num -= 3
    count += 1

else:
    print(-1)
```
